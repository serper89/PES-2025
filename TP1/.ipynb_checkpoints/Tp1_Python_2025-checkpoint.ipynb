# Ejercicio 1 - Filtro Moving Average

1. Implementá un algoritmo de promedio móvil (moving average) según la definición vista en clase, que reciba como parámetros:

- La señal de entrada.
- El tamaño de la ventana.

A partir de tu implementación, obtené y graficá la respuesta impulsiva del sistema (lo que en clase llamamos su “firma” o función de transferencia).

2. Respondé:

- ¿Cómo se obtiene la respuesta impulsiva de cualquier sistema?
- ¿Qué representa en el caso del filtro de promedio móvil?

```python
import scipy.signal as sig
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft

"""armamos la funcion como parametro de entrada la señal y el tamaño de la ventama"""
def mediamovil(señal, ventana):
    """armo el vector que me almacena el promedio de la señal"""
    señalpromedio = np.zeros(len(señal))                             
    """bucle for que barre la señal"""
    for n in range(len(señal)):                                      
        """inicializo la variable suma del promedio"""    
        suma = 0
        """inicializo la variable cantidad del promedio"""
        d = 0
        """el promedio se hace sobre una ventana fija del conjunto dominio de la señal"""
        for k in range (-(ventana), ventana):                      
            """condicional para que se haga el promedio dentro del dominio de la señal"""
            if ((n-k) > 0) and ((n-k) < (len(señal))):       
                suma = suma + señal[(n-k)]
                d = d+1
                """se almacena el promedio en casa muestreo de la señal"""
                señalpromedio[n] = np.copy(suma/d)               
    """se grafica la señal promediada"""
    #plt.plot(t,señalpromedio)                                                  
    return señalpromedio

    
########### Parámetros Iniciales ###########
Fs = 10000   # Sampling Freq 10KHz
N = 200      # Samples number
t = np.linspace(0,N/Fs,num=N) # Crea una variable de tiempo desde 0, con intervalos de 1/Fs

f1 = 150   # 100Hz
A1 = 3

signal = A1 * np.sin(2*np.pi*f1*t)

noise = np.random.normal(0,1,len(signal))
sig_noise = signal + noise
plt.plot(t,sig_noise)
señalfiltrada = mediamovil(sig_noise, 21)
plt.plot(t,señalfiltrada)
```

```python
f = np.linspace(-Fs/2,Fs/2,N)

spec_sig_noise = abs(np.fft.fftshift(fft(sig_noise)))/N
spec_señalfiltrada = abs(np.fft.fftshift(fft(señalfiltrada)))/N

plt.title("Espectro Señal+Ruido")
plt.stem(f,(spec_sig_noise))
plt.show()
plt.title("Espectro Señal filtrada")
plt.stem(f,spec_señalfiltrada)
```

```python
# Ejemplos de medicion de tiempos
# %pylab inline
import timeit
import time
import numpy as np
import matplotlib.pyplot as plt
from decimal import Decimal, getcontext
import math
from scipy.signal import convolve
from scipy.fft import fft, fftshift, fftfreq, ifft

t = np.arange(0, 1024/20000, 1/20000)
s1= np.sin(2*np.pi*1000*t)+1
s2 = np.sin(2*np.pi*1200*t)

"""Hacemos la precision de 3 decimales de punto fijo y convertimos las señales a punto fijo para posterior ser implementadas"""
getcontext().prec = 3
s1_fijo=np.zeros(len(s1))
s2_fijo=np.zeros(len(s2))
conv_fijo=np.zeros(len(s1)+len(s2)-1)
for n in range(len(s1)):
    s1_fijo[n]=+Decimal(s1[n])
for m in range(len(s2)):
    s2_fijo[m]=+Decimal(s2[m])
for l in range (len(conv_fijo)):
    conv_fijo[l]=+Decimal(conv_fijo[l])

#print (s1_fijo)
#print (s2_fijo)
#print (conv_fijo)

#Implementamos algoritmo de convolucion de punto fijo
def convolucion_fijo():
    """bucle for que barre la señal1"""
    for n in range(len(s1_fijo)):                                      
        """bucle for que barre la señal1"""
        for k in range (len(s2_fijo)):                      
            """Convolucionamos por definicion"""
            conv_fijo[n+k]=conv_fijo[n+k]+(s1_fijo[n]*s2_fijo[k])                                                 
    return



#Implementamos algoritmo de convolucion con python puro
"""armamos la convolucion como parametro de entrada las señales"""
def convolucion_puro(señal1, señal2):
    """Armamos el vector(lista) de salida de la convolucion"""
    convolucion=np.zeros(len(señal1)+len(señal2)-1)
    """bucle for que barre la señal1"""
    for n in range(len(señal1)):                                      
        """bucle for que barre la señal1"""
        for k in range (len(señal2)):                      
            """Convolucionamos por definicion"""
            convolucion[n+k]=convolucion[n+k]+(señal1[n]*señal2[k])                                                 
    return  convolucion



# METODO 2 de medición de tiempos
print ("Tiempo usando nunpy")
start = time.time()
conv_r1 = np.convolve(s1,s2)
end = time.time()
print (end-start)
print (conv_r1)
print ("Tiempo usando python puro (punto flotante)")
start = time.time()
conv_r2=convolucion_puro(s1,s2)
end = time.time()
print (end-start)
print (conv_r2)
print ("Tiempo usando python puro (punto fijo)")
start = time.time()
convolucion_fijo()
end = time.time()
print (end-start)
print (conv_fijo)


#plt.plot(np.arange(len(resultado_conv))/1000, resultado_conv)
#t = np.arange(0, (len(s1)+len(s2)-1)/20000, 1/20000)
#t = np.arange(0, (len(s1)+len(s2)-1), 1/20000)
plt.figure(figsize=(12, 24))
plt.subplot(6, 1, 1)
plt.title('Convolucion Numpy')
plt.plot(np.arange(len(conv_r1))/20000, conv_r1)
plt.subplot(6, 1, 2)
plt.title('Convolucion Punto Flotante')
plt.plot(np.arange(len(conv_r2))/20000, conv_r2)
plt.subplot(6, 1, 3)
plt.title('Convolucion Punto Fijo')
plt.plot(np.arange(len(conv_fijo))/20000, conv_fijo)

espectro_s1 = (fft(s1))
espectro_s2 = (fft(s2))
espectro_conv = espectro_s1 * espectro_s2
freqs = fftshift(fftfreq(len(t), 1/20000))

plt.subplot(6, 1, 4)
plt.stem(freqs, np.abs(fftshift(espectro_s1)))
plt.title('Espectro del Señal 1')
plt.xlabel('Frecuencia')
plt.ylabel('Amplitud')

plt.subplot(6, 1, 5)
plt.stem(freqs, np.abs(fftshift(espectro_s2)))
plt.title('Espectro del Señal 2')
plt.xlabel('Frecuencia')
plt.ylabel('Amplitud')

plt.subplot(6, 1, 6)
plt.stem(freqs, np.abs(fftshift(espectro_conv)))
plt.title('Espectro del Resultado de la Convolución')
plt.xlabel('Frecuencia')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()

```
