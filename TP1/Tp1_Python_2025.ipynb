# Ejercicio 1 - Filtro Moving Average

1. Implementá un algoritmo de promedio móvil (moving average) según la definición vista en clase, que reciba como parámetros:

- La señal de entrada.
- El tamaño de la ventana.

A partir de tu implementación, obtené y graficá la respuesta impulsiva del sistema (lo que en clase llamamos su “firma” o función de transferencia).

2. Respondé:

- ¿Cómo se obtiene la respuesta impulsiva de cualquier sistema?
- ¿Qué representa en el caso del filtro de promedio móvil?

```python
import scipy.signal as sig
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft

"""armamos la funcion como parametro de entrada la señal y el tamaño de la ventama"""
def mediamovil(señal, ventana):
    """armo el vector que me almacena el promedio de la señal"""
    señalpromedio = np.zeros(len(señal))                             
    """bucle for que barre la señal"""
    for n in range(len(señal)):                                      
        """inicializo la variable suma del promedio"""    
        suma = 0
        """inicializo la variable cantidad del promedio"""
        d = 0
        """el promedio se hace sobre una ventana fija del conjunto dominio de la señal"""
        for k in range (-(ventana), ventana):                      
            """condicional para que se haga el promedio dentro del dominio de la señal"""
            if ((n-k) > 0) and ((n-k) < (len(señal))):       
                suma = suma + señal[(n-k)]
                d = d+1
                """se almacena el promedio en casa muestreo de la señal"""
                señalpromedio[n] = np.copy(suma/d)               
    """se grafica la señal promediada"""
    #plt.plot(t,señalpromedio)                                                  
    return señalpromedio

    
########### Parámetros Iniciales ###########
Fs = 10000   # Sampling Freq 10KHz
N = 200      # Samples number
t = np.linspace(0,N/Fs,num=N) # Crea una variable de tiempo desde 0, con intervalos de 1/Fs

f1 = 150   # 100Hz
A1 = 3

signal = A1 * np.sin(2*np.pi*f1*t)

noise = np.random.normal(0,1,len(signal))
sig_noise = signal + noise
plt.plot(t,sig_noise)
señalfiltrada = mediamovil(sig_noise, 21)
plt.plot(t,señalfiltrada)
```

```python
f = np.linspace(-Fs/2,Fs/2,N)

spec_sig_noise = abs(np.fft.fftshift(fft(sig_noise)))/N
spec_señalfiltrada = abs(np.fft.fftshift(fft(señalfiltrada)))/N

plt.title("Espectro Señal+Ruido")
plt.stem(f,(spec_sig_noise))
plt.show()
plt.title("Espectro Señal filtrada")
plt.stem(f,spec_señalfiltrada)
```
